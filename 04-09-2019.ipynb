## Functions

```python
#program to find leap year
#functions is python are defined by a keyword "def"
#leap year program
def isleapyear(year):
    if (year%400==0) or (year%4==0 and year%100!=0):
        print("leaap year")
    else:
        print("not a leap year")
isleapyear(2020)        
```

```python
isleapyear(2020)
```
